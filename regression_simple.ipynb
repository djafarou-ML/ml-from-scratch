```python
import matplotlib.pyplot as plt

# Données
x = [1, 2, 3, 4, 5]
y = [180, 220, 250, 310, 360]
m = len(x)

# Paramètres
w = 40
b = 100
alpha = 0.03
iterations = 10

# Historique
w_history = []
b_history = []
cost_history = []

# Descente de gradient
for i in range(iterations):
    y_pred = []
    errors = []
    
    # Prédictions
    for j in range(m):
        yp = w * x[j] + b
        y_pred.append(yp)
        errors.append(yp - y[j])
    
    # Gradients
    dw = sum(errors[j] * x[j] for j in range(m)) / m
    db = sum(errors) / m
    
    # Mise à jour avec nouvelles variables
    w_new = w - alpha * dw
    b_new = b - alpha * db
    
    # Remplacement
    w = w_new
    b = b_new
    
    # Coût
    cost = sum(e**2 for e in errors) / (2*m)
    
    w_history.append(w)
    b_history.append(b)
    cost_history.append(cost)
    
    print(f"Iteration {i+1}: w={w:.2f}, b={b:.2f}, cost={cost:.2f}")

# Graphiques
plt.figure(figsize=(12,5))

# Droites de régression
plt.subplot(1,2,1)
plt.scatter(x, y, label='Données')

x_line = [min(x), max(x)]

for i in range(iterations):
    y_line = [w_history[i]*xi + b_history[i] for xi in x_line]
    plt.plot(x_line, y_line, alpha=0.3)

# Dernière droite
y_final = [w_history[-1]*xi + b_history[-1] for xi in x_line]
plt.plot(x_line, y_final, linewidth=2, label='Droite finale')

plt.title("Évolution de la droite de régression")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()

# Courbe du coût
plt.subplot(1,2,2)
plt.plot(range(1, iterations+1), cost_history, marker='o')

plt.title("Descente de gradient")
plt.xlabel("Itérations")
plt.ylabel("J(w,b)")

plt.tight_layout()
plt.show()
```
